```python
target_height = 224  # Example height
target_width = 224   # Example width

import tensorflow as tf

def resize_image(image_path, target_height, target_width):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_image(image, channels=3)
    image = tf.image.resize(image, [target_height, target_width])
    image = tf.cast(image, tf.uint8)
    return image

# Define target dimensions
target_height = 224  # Replace with your desired height
target_width = 224   # Replace with your desired width

# Resize the image
resized_image = resize_image(new_image_path, target_height, target_width)
```
